```python
raw_data=pd.read_csv('/kaggle/input/playground-series-s4e7/train.csv')
df=raw_data.copy()
df.head()
```

```python
test = pd.read_csv('/kaggle/input/playground-series-s4e7/test.csv')
```

```python
print(df[['Gender', 'Vehicle_Age', 'Vehicle_Damage', 'Driving_License', 'Previously_Insured']].isna().sum())
print(test[['Gender', 'Vehicle_Age', 'Vehicle_Damage', 'Driving_License', 'Previously_Insured']].isna().sum())
```

```python
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1}).fillna(0).astype(int)
df['Vehicle_Age'] = df['Vehicle_Age'].map({'< 1 Year': 0, '1-2 Year': 1, '> 2 Years': 2}).fillna(0).astype(int)
df['Vehicle_Damage'] = df['Vehicle_Damage'].map({'No': 0, 'Yes': 1}).fillna(0).astype(int)
df['Driving_License'] = df['Driving_License'].fillna(0).astype(int)
df['Previously_Insured'] = df['Previously_Insured'].fillna(0).astype(int)

test['Gender'] = test['Gender'].map({'Female': 0, 'Male': 1}).fillna(0).astype(int)
test['Vehicle_Age'] = test['Vehicle_Age'].map({'< 1 Year': 0, '1-2 Year': 1, '> 2 Years': 2}).fillna(0).astype(int)
test['Vehicle_Damage'] = test['Vehicle_Damage'].map({'No': 0, 'Yes': 1}).fillna(0).astype(int)
test['Driving_License'] = test['Driving_License'].fillna(0).astype(int)
test['Previously_Insured'] = test['Previously_Insured'].fillna(0).astype(int)
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score

models = [GaussianNB(), # Naive Bayes Model
          DecisionTreeClassifier(random_state=42), # Decision Tree Model
          RandomForestClassifier(n_estimators=100, random_state=42), # Random forest model
          LogisticRegression(random_state=50), # Logistic regression model
          AdaBoostClassifier(random_state=45), # Ada Boost Model
         ]
          

for i, model in enumerate(models):
    model.fit(X_train, y_train)
    train_accuracy = accuracy_score(y_train, model.predict(X_train))
    test_accuracy = accuracy_score(y_test, model.predict(X_test))
    print(model)
    print()
    print(f"Model {i+1}: {type(model).__name__}")
    print()
    print(f"Training Accuracy: {train_accuracy}")
    print(f"Testing Accuracy: {test_accuracy}")
    print("------------------")
```

From the above we can clearly see that there is overfitting case in Decision Tree and Random Forest models. Whereas GaussianNB, LogisticRegression, and AdaBoostClassifier all have similar training and testing accuracies, indicating they generalize better than the other models and are less likely to be overfitting.
